# linear graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r'
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,7]
y = [4,1,6,9]
plt.plot(x,y)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y, marker = 'o')
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
plt.plot(y, marker = '>')
plt.show()
```

# bar graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r','g','b','y'
plt.bar(x,y, color = c)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.barh(x,y, color = 'hotpink')
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.bar(x,y, color = 'red', width = 0.1)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.barh(x,y, color = 'red',  height = 0.1)
plt.show()
```
